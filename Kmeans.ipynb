```python
import tensorflow as tf

# 创建空列表，用于存储数据
radical_r_list = []
radical_g_list = []
seed_r_list = []
seed_g_list = []

# 读取第一个 CSV 文件的数据
with open('/content/Bayes_data_radical.csv', 'r') as file:
    # 逐行读取文件内容
    for i, line in enumerate(file):
        if i >= 250:
            break
        # 将每一行按逗号分割并取第一个元素和第二个元素，转换为浮点数类型，并添加到列表中
        radical_r = float(line.strip().split(',')[0])
        radical_g = float(line.strip().split(',')[1])
        radical_r_list.append(radical_r)
        radical_g_list.append(radical_g)

# 读取第二个 CSV 文件的数据
with open('/content/Bayes_data_seed.csv', 'r') as file:
    # 逐行读取文件内容
    for i, line in enumerate(file):
        if i >= 250:
            break
        # 将每一行按逗号分割并取第一个元素和第二个元素，转换为浮点数类型，并添加到列表中
        seed_r = float(line.strip().split(',')[0])
        seed_g = float(line.strip().split(',')[1])
        seed_r_list.append(seed_r)
        seed_g_list.append(seed_g)

# 将列表转换为 TensorFlow 张量
radical_r_tensor = tf.convert_to_tensor(radical_r_list, dtype=tf.float64)
radical_g_tensor = tf.convert_to_tensor(radical_g_list, dtype=tf.float64)
seed_r_tensor = tf.convert_to_tensor(seed_r_list, dtype=tf.float64)
seed_g_tensor = tf.convert_to_tensor(seed_g_list, dtype=tf.float64)

# 将 radical_r 和 seed_r 合并成一个张量
combined_r_tensor = tf.concat([radical_r_tensor, seed_r_tensor], axis=0)

# 将 radical_g 和 seed_g 合并成一个张量
combined_g_tensor = tf.concat([radical_g_tensor, seed_g_tensor], axis=0)

# 输出合并后的张量
print("Combined Radical R and Seed R:", combined_r_tensor)
print("Combined Radical G and Seed G:", combined_g_tensor)

# 将 radical_r_list 和 radical_g_list 合并成一个张量
combined_tensor_rg = tf.stack([combined_r_tensor, combined_g_tensor], axis=1)

# 输出合并后的张量
print(combined_tensor_rg)
```

```python
import sklearn.datasets as skd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
numFeatures           = 2
numClusters           = 2
numSamples            = 500
numIter               = 20
centroids = combined_tensor_rg
samples, labels = skd.make_blobs(n_samples = numSamples, centers = centroids, cluster_std = 0.9)
points    = samples
centroids = tf.convert_to_tensor(centroids)
labels    = tf.convert_to_tensor(labels)
samples   = tf.convert_to_tensor(samples)
```

```python
kmeans = tf.compat.v1.estimator.experimental.KMeans(num_clusters = numClusters, use_mini_batch = False)
previousCenters = None
for _ in range(numIter):
  kmeans.train(input_fn)
  clusterCenters = kmeans.cluster_centers()
  if previousCenters is not None:
    print('Cluster centers variation:', clusterCenters - previousCenters)
  previousCenters = clusterCenters
  print('Algorithm score:', kmeans.score(input_fn))
print('Cluster centers:', clusterCenters)
clusterCenters = kmeans.cluster_centers()
clusterLabels = list(kmeans.predict_cluster_index(input_fn))
for i, point in enumerate(points):
    clusterIndex  = clusterLabels[i]
    center        = clusterCenters[clusterIndex]
    #print('point:', point, 'is in cluster', clusterIndex, 'centered at', center)

plotClusters(samples, clusterLabels, clusterCenters)
```

```python
def plotClusters(samples, clusterLabels, clusterCenters):
    colors = ['red', 'blue', 'green', 'yellow', 'purple', 'orange', 'brown', 'pink', 'gray', 'cyan']  # 擴展顏色列表

    for i, centroid in enumerate(clusterCenters):
      # Convert clusterLabels to numpy array for proper indexing
      clusterLabels_array = np.array(clusterLabels)
      # Ensure each cluster label corresponds to a color in the list
      cluster_colors = [colors[label] for label in clusterLabels_array]
      plt.scatter(samples[:, 0], samples[:, 1], c=cluster_colors,s =1, label=f'Cluster {i}')

    plt.scatter(clusterCenters[:, 0], clusterCenters[:, 1], c='black', marker='x', s=100, label='Centroids')
    # Create a custom legend with unique labels and colors for each cluster
    unique_labels = np.unique(clusterLabels)
    custom_legend = [plt.Line2D([0], [0], marker='o', color='w', markersize=10, markerfacecolor=colors[label], label=f'Cluster {label}') for label in unique_labels]
    plt.legend(handles=custom_legend)
    plt.xlabel('R')
    plt.ylabel('G')
    plt.show()
```

```python
def input_fn():
  return tf.compat.v1.train.limit_epochs(tf.convert_to_tensor(points, dtype = tf.float32), num_epochs = 1)
```
